CREATE VOCABULARY FIRST

```python
df = pd.read_csv('spam.csv')
```

```python
df.head()
```

```python
df['type'].value_counts()
```

```python
df.drop(['title'],inplace=True,axis=1)
```

```python
df.head()
```

```python
df['spam or not'] = df['type'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(df['text'],df['spam or not'],test_size=0.2)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
x_train_cv = v.fit_transform(xtrain.values)
```

```python
x_train_cv.toarray()[:1].shape
```

```python
v.get_feature_names_out()[200:250]
```

```python
dir(v)
```

```python
v.vocabulary_
```

```python
x_train_np = x_train_cv.toarray()
```

```python
x_train_np[:4]
```

```python
indices = np.where(x_train_np[0]!=0)
```

```python
for i in indices:
    print(x_train_np[0,i])
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train_cv,ytrain)
```

```python
xtest_cv = v.transform(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = model.predict(xtest_cv)
```

```python
print(classification_report(ytest,y_pred))
```
